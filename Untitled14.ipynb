```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('householdtask3.csv')
data
```

```python
data.head(10)
```

```python
#scatter plot
colors = np.random.rand(70)
plt.scatter(data['year'],data['own'],c=colors,alpha=0.9,cmap='viridis')
plt.colorbar()
plt.title('scatter plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()            
```

```python
#line plot
plt.plot(data['year'],data['income'],color='b',linestyle=':',marker='o')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.title('lineplot')
plt.xlabel('year')
plt.ylabel('income')
plt.show()
```

```python
#bar plot
plt.bar(data['year'],data['eqv_income'],color='r',width=1.5,align='center',edgecolor='y')
plt.title('BarPlot')
plt.xlabel('year')
plt.ylabel('eqv_income')
plt.show()
```

```python
# HORIXONTAL bar plot
plt.barh(data['year'],data['eqv_income'])
plt.title('BarPlot')
plt.xlabel('year')
plt.ylabel('eqv_income')
plt.show()
```

```python
#stacked bar plot
plt.bar(data['year'],data['income'],color='green',width=1.5)
plt.bar(data['year'],data['eqv_income'],color='blue',width=1.5)
plt.title('BarPlot')
plt.xlabel('year')
plt.ylabel('eqv_income and income')
plt.show()
```

```python
plt.hist(data['income'],color='yellow',edgecolor='r')
plt.hist(data['eqv_income'],color='m',edgecolor='g')
plt.title('HISTOGRAM')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```
