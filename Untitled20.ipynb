1.HOW MANY PEOPLE HAVE HEART DISEASE AND HOW MANY DONT HAVE HEART DISEASE

```python
df.target.value_counts()
```

```python
ax=df.target.value_counts().plot(kind='bar',color=['orchid','orange'],figsize=(8,6))
plt.xlabel('target')
plt.title('analysis')
plt.legend(['disease','No disease'])
plt.show()
```

```python
df.target.value_counts().plot(kind='pie',)
plt.title('ANALYSES')
plt.legend(['disease','No disease'])
plt.show()
```

2.PEOPLE OF WHICH SEX HAVE HEART DISEASE?

```python
df.sex.value_counts()
```

```python
color=['m','y']
df.sex.value_counts().plot(kind='pie',colors=color)
plt.title('MALE FEMALE RATIO')
plt.legend(['MALE','FEMALE'])
plt.show()
```

```python
pd.crosstab(df.target,df.sex,margins=True,margins_name='Total')
```

```python
df['sex'] = df['sex'].astype(str)

sns.countplot(x='target',data=df,hue='sex')
plt.title('HEART DISEASE FREQUENCY FOR SEX')
plt.xlabel('0=no heart disease,1=heart disease')
```

3.PEOPLE OF WHICH SEX HAS WHICH TYPE OF CHEST PAIN MOST?

```python
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title('chest pain type vs count')
```

```python
pd.crosstab(df.cp,df.sex)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=['coral','plum','khaki','c'])
plt.title('TYPE OF CHEST PAIN VS SEX')
plt.xlabel('0=female, 1=male')
```

6.PEOPLE OF WHICH AGE HAS MOST CA?

```python
df.ca.value_counts()
```

```python
df.ca.value_counts().plot(kind='bar',color='pink')
plt.title('CA ANALYSES')
plt.xticks(rotation=45, ha='right')
```

```python
pd.crosstab(df.ca,df.age)
```

```python
pd.crosstab(df.age,df.ca).plot(kind='bar',color=['coral','c','pink','yellow'],figsize=(15,12),width=1)
plt.title('AGE VS CA')
plt.xlabel('0=female, 1=male')
```
